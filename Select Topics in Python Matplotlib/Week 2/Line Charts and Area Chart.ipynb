Creating Line Charts

```python
import matplotlib.pyplot as plt
```

```python
days=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
max_temp=[65,50,20,34,43,40,57]

plt.plot(days,max_temp)
plt.title('Temperature this week')
plt.xlabel('days')
plt.ylabel('max_temp')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
a= random.sample(range(50), 20)
b= random.sample(range(50), 20)
c=random.sample(range(25), 20)
plt.plot(a,label= "line 1" )
plt.plot(b, color="green")
plt.plot(c, color="red",label="line 3",linestyle=":")
plt.legend(loc="upper left")
plt.show()

```

Creating an area Charts

```python
groups2= ['freshman','sophmore','junior','senior']
total=[40,31,15,14]
total2=[60,40,33,18]
```

```python
plt.fill_between(groups2, total,total2)
plt.show()
```

```python
plt.stackplot(groups2, total,total2)
plt.show()
```
