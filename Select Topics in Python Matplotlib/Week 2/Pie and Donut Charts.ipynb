Creating Pie Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline

groups2=['male','female']

number=[25,25]

# Creating plot

plt.pie(number, labels = groups2)

# show plot
plt.show()
```

```python
groups2= ['freshman','sophomore','junior','senior']
total=[40,31,15,14]
plt.pie(total, labels = groups2, autopct='%1.1f%%')
plt.show()
```

```python
groups2= ['freshman','sophomore','junior','senior']
total=[140,61,35,20]
plt.pie(total, labels = groups2, autopct='%1.1f%%')
plt.show()
```

Creating Donut Charts

```python
groups2= ['freshman','sophomore','junior','senior']
total=[40,31,15,14]
# Creating plot

plt.pie(total, labels = groups2, wedgeprops=dict(width=0.3),autopct='%1.1f%%')
# show plot
plt.show()
```

```python
plt.pie(total, labels = groups2) 
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
```

Multiple Charts

```python
groups2= ['freshman','sophomore','junior','senior']
total=[140,31,25,10]
total2=[150,20,16,12]
```

```python
plt.figure(0)
plt.pie(total, labels = groups2, autopct='%1.1f%%')
plt.figure(2)
plt.pie(total2, labels = groups2, autopct='%1.1f%%')
plt.show()

```

```python
groups2= ['freshman','sophomore','junior','senior']
total=[140,31,25,10]
total2=[150,20,16,12]
total3=[12,32,42,12]
plt.figure(1)
plt.pie(total, labels = groups2, autopct='%1.1f%%')
plt.figure(2)
plt.pie(total2, labels = groups2, autopct='%1.1f%%')
plt.figure(1)
plt.pie(total3, labels = groups2, autopct='%1.1f%%')
plt.show()
```
