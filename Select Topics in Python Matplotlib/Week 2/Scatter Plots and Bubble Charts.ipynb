Creating Scatter Plots

```python
import matplotlib.pyplot as plt
import random
x = random.sample(range(100), 20)
y = random.sample(range(100), 20)
print(x)
print(y)
```

```python
x=[27, 32, 38, 94, 70, 29, 17, 8, 48, 82, 52, 14, 91, 22, 58, 96, 73, 99, 75, 76]
y=[78, 35, 28, 75, 2, 22, 25, 17, 72, 45, 86, 96, 24, 41, 73, 51, 58, 76, 90, 77]
plt.scatter(x, y)
plt.title('title name')
plt.xlabel('xAxis name')
plt.ylabel('yAxis name')
plt.show()

plt.show()
```

Bubble Charts

```python
# create data
import numpy as np
x = random.sample(range(100), 20)
y = random.sample(range(100), 20)
```

```python
plt.scatter(x, y)
```

```python
# create data
x = random.sample(range(100), 20)
y = random.sample(range(100), 20)
z = random.sample(range(1000), 20)
 
# use the scatter function to create our bubble chart
plt.scatter(x, y, s=z , alpha=0.5)

plt.title('title name')
plt.xlabel('xAxis name')
plt.ylabel('yAxis name')
plt.show()


```
